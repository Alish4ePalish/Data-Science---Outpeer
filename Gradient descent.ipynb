# Gradient Descent

```python
def gradient_descent(f_prime, start_x, learning_rate, iterations):
    """
    Parameters:
    - f_prime: derivative (gradient)
    - start_x: starting point
    - learning_rate: rate of descent

    Returns:
    - x: optimal quantity.
    - history: list of x values at each step
    """
    x = start_x
    history = [x]
    for _ in range(iterations):
        gradient = f_prime(x)
        x -= learning_rate * gradient  # Обновляем x
        history.append(x)
    return x, history

# Function: f(x) = x^2 - 40x + 400
# Derivative: f'(x) = 2x - 40
def f_prime(x):
    return 2 * x - 40

# Parameters for gradient descent
start_x = 0  
learning_rate = 0.1  
iterations = 50  

# Run gradient descent
optimal_x, history = gradient_descent(f_prime, start_x, learning_rate, iterations)

print(f"Optimal quantity: {optimal_x}")
```

```python
import matplotlib.pyplot as plt

# Graph of change of X at each step
plt.plot(history, marker='o')
plt.title('Gradient descent process')
plt.xlabel('Iteration')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()
```
